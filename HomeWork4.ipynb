```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

Изучите стоимости недвижимости


```python
plt.figure(figsize=(8, 8))
plt.hist(df['price'])
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во')
```

Изучите распределение квадратуры жилой


```python
plt.figure(figsize=(6, 12))
plt.hist(df['sqft_living'])
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры')
plt.ylabel('Количество')
```

Изучите распределение года постройки


```python
plt.figure(figsize=(6, 8))
sns.histplot(df['yr_built'], bins=20)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```

### 2 задача

Изучите распределение домов от наличия вида на набережную. 
Постройте график, сделайте выводы


```python
plt.figure(figsize=(6, 6))

plt.pie(df['waterfront'].value_counts(), autopct='%1.1f%%')
plt.legend(['Без вида на набережную', 'С видом на набережную'])
```

Вывод: менее одного процента домов имеют вид на набережную.

Изучите распределение этажей домов


```python
data = df['floors'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(data, autopct='%1.2f%%', labels=data.index)
plt.title('Количество этажей в доме')
```

Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

### 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'))
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Зависимость стоимости от кол-ва этажей');
```

Цена зависит от кол-ва этажей.
Выше всего цена там, где 2,5

```python
sns.jointplot(x=df['price'], y=df['sqft_lot'], kind='scatter');
```

Цена зависит от общей площади дома. Дома с большей площадью дороже.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Цена зависит от качества постройки. Чем выше качество, тем выше цена.

```python
matrix = df.corr(numeric_only = True)
matrix = np.round(matrix, 2)
matrix

plt.figure(figsize=(10, 10))
sns.heatmap(matrix, annot = True, cmap ='coolwarm');
```

Наибольшие зависимости по цене от: общей площади, жилой площади и качества дома.
